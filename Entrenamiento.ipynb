```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")   # modelo peque√±o

model.train(
    data="/content/drive/MyDrive/datasetP600/data.yaml",

    epochs=20,               # ajusta seg√∫n tu GPU
    batch=16,
    imgsz=640,
    device=0                 # usa GPU
)
```

```python
import shutil
import os

def guardar_modelo(nombre):
  source = "runs/detect/train/weights/best.pt"     # Trained model

  destination_folder = "/content/drive/MyDrive/datasetP/modelo"
  os.makedirs(destination_folder, exist_ok=True)

  destination = f"/content/drive/MyDrive/datasetP/modelo/{nombre}.pt"  # Your target folder

  shutil.copy(source, destination)
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def predecir(ruta_img , predict_folder):
  results = model.predict(
      source=ruta_img, save=True
  )

  # obtener cajas
  boxes = results[0].boxes

  # coordenadas xyxy
  xyxy = boxes.xyxy.cpu().numpy()
  cls = boxes.cls.cpu().numpy().astype(int)

  # tama√±o de la imagen
  img = Image.open(ruta_img)
  w_img, h_img = img.size

  yolo_lines = []

  for i in range(len(boxes)):
      x1, y1, x2, y2 = xyxy[i]

      # convertir xyxy ‚Üí xywh
      x_center = (x1 + x2) / 2.0
      y_center = (y1 + y2) / 2.0
      width = (x2 - x1)
      height = (y2 - y1)

      # normalizar
      x_center /= w_img
      y_center /= h_img
      width /= w_img
      height /= h_img

      line = f"{cls[i]} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}"
      yolo_lines.append(line)


  os.makedirs(predict_folder, exist_ok=True)

  # Nombre del archivo de salida
  txt_file = os.path.join(predict_folder, os.path.splitext(os.path.basename(ruta_img))[0] + ".txt")

  # Guardar predicciones
  with open(txt_file, "w") as f:
      for l in yolo_lines:
          f.write(l + "\n")
```

```python
import glob
import shutil
import os

carpeta_val = "/content/drive/MyDrive/datasetP600/images/train"
predict_folder = "/content/drive/MyDrive/datasetP600/predictTrain"
imagenes = glob.glob(os.path.join(carpeta_val, "*.jpg"))     # all JPGs

print("Total images:", len(imagenes))

for img_path in imagenes:
    print("Predicting:", img_path)
    predecir(img_path,predict_folder)
```
